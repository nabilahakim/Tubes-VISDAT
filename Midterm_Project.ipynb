### Soal 1 (15 poin)
Tampilkan visualisasi distribusi data untuk variabel "Glucose" dan "Insulin", lalu jelaskan hasil distribusinya.

#### Code

```python
# type your code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/diabetes (1).csv')
df.head()
```

```python
db = df[['Glucose', 'Insulin']]
db.head()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
db.index = db.index.map(int)

db['Insulin'].plot(kind='hist', figsize=(20,6), ax=ax1, color='blue')
db['Glucose'].plot(kind='hist', figsize=(20,6), ax=ax1, color='green')
ax1.set_title ('Plot Grafik Distribusi Glucose dan Insulin')
ax1.set_ylabel('Number of People')
ax1.set_xlabel('Value')

plt.show()
```

#### Penjelasan

Dalam visualisasi data diatas ditampilkan hasil distribusi berdasarkan nilai glukosa dan insulin yang dipunyai oleh setiap orang. Dapat dilihat pada setiap nilai glukosa dan insulin meningkat jika nilai tersebut dimiliki oleh banyak orang dan begitu sebaliknya jika nilai dari keduanya rendah.

### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
# type your code here
def trend2019(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
```

## Hang_seng dataset

```python
df_hang = pd.read_csv('/content/Hang_Seng.csv')
df_hang = df_hang.round(4)
df_hang.head()
```

```python
df_hang['Date'] = pd.to_datetime(df_hang['Date'])
df_hang['year'] = df_hang['Date'].dt.year
df_hang.head()
```

```python
df_hang["Day_Perc_Change"] = df_hang["Adj Close"].pct_change()*100
df_hang["trend2019"]= df_hang["Day_Perc_Change"].apply(lambda x:trend2019(x))
df_hang.head()
```

```python
df_hang.index = df_hang["year"]
df_hang2019 = df_hang.loc['2019']
df_hang2019
```

```python
plt.subplots(figsize = (15,10))
plt.title('Trend Analysis Hang Seng 2019')
ax = sns.countplot(y='trend2019', data=df_hang2019)


plt.show()
```

## Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# type your code here
df_hang2019 = df_hang2019.reset_index(drop=True)
df_nasdaq2019 = df_nasdaq2019.reset_index(drop=True)
df_nikkei2019 = df_nikkei2019.reset_index(drop=True)

df_comb = pd.concat([df_hang2019["Adj Close"], df_nasdaq2019["Adj Close"], df_nikkei2019["Adj Close"]], axis=1)
df_comb.columns = ["Hang Seng", "Nasdaq", "Nikkei"]

df_comb.dropna(inplace= True, axis =0)
df_comb.head()
```

```python
pct_chg_df = df_comb.pct_change()*100
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)
volatility = pct_chg_df[["Hang Seng", "Nasdaq", "Nikkei"]].rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (15, 7))
```

#### Penjelasan
Pada 3 dataset saham dari perusahaan yang berbeda dilakukan penggabungan dalam satu dataframe untuk dicari nilai volatilitas berdasarkan nilai adj close yang ada di setiap dataset saham tersebut. Dari grafik diatas bisa dilihat nilai volatilitas dari saham nasdaq memilki nilai volatilitas tertinggi dibanding dengan saham yang lainnya sedangkan dua saham lainnya nilai volatilitas tertingginya hampir sama dengan yang lainnya.
